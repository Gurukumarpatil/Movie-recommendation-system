```python
# ==============================
# 1. IMPORTS
# ==============================
import pandas as pd
import numpy as np

# ==============================
# 2. LOAD DATA
# ==============================
movies = pd.read_csv('../data/raw/movies.csv')
ratings = pd.read_csv('../data/raw/ratings.csv')

# ==============================
# 3. BASIC INSPECTION
# ==============================
print("Movies Data:")
print(movies.head())

print("\nRatings Data:")
print(ratings.head())

# ==============================
# 4. DATA INFO
# ==============================
print("\nMovies Info:")
print(movies.info())

print("\nRatings Info:")
print(ratings.info())

# ==============================
# 5. UNIQUE COUNTS
# ==============================
n_users = ratings['userId'].nunique()
n_movies = ratings['movieId'].nunique()

print(f"\nNumber of Users: {n_users}")
print(f"Number of Movies: {n_movies}")

# ==============================
# 6. CREATE USER-ITEM MATRIX
# ==============================
user_item = ratings.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
)

print("\nUser-Item Matrix Shape:")
print(user_item.shape)

# ==============================
# 7. SPARSITY CHECK
# ==============================
missing_values = user_item.isna().sum().sum()
total_values = user_item.size

sparsity = (missing_values / total_values) * 100

print(f"\nTotal Missing Values: {missing_values}")
print(f"Sparsity: {sparsity:.2f}%")

# ==============================
# 8. SAMPLE MATRIX
# ==============================
print("\nSample User-Item Matrix:")
print(user_item.head())

# ==============================
# 9. SAVE MATRIX
# ==============================
user_item.to_csv('../data/processed/user_item_matrix.csv')

print("\nUser-item matrix saved successfully!")
```
